```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Prepare dataset
data = np.loadtxt('/content/drive/MyDrive/DataScienceLab/정규 세션/220210_정규세션_Ensemble/data.csv', delimiter=',', dtype=float)
labels = data[:, 0]
features = preprocessing.minmax_scale(data[:, 1:])
X_train, X_test, y_train, y_test = train_test_split(features, labels.ravel(), test_size=0.3)
```

### 의사결정나무
* random_state = 2022 으로 설정
* 변수명은 dt_clf 로 설정

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# 빈 부분 코드 작성
# decision tree
dt_clf = DecisionTreeClassifier(random_state = 2022)
```

```python
# 개별 분류기에 train set 피팅
dt_clf.fit(X_train,y_train)
```

```python
# test셋으로 prediction
dt_pred = dt_clf.predict(X_test)
```

```python
# 성능 확인
# accuracy_score 계산
print(accuracy_score(y_test, dt_pred))
```

### 랜덤 포레스트
* random_state = 2022
* 변수명 rf_clf 로

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# random forest
rf_clf = RandomForestClassifier(random_state = 2022)
```

```python
# 개별 분류기에 train set 피팅
rf_clf.fit(X_train, y_train)
```

```python
# test셋으로 prediction
rf_pred = rf_clf.predict(X_test)
```

```python
# 성능 확인
# accuracy_score 계산
print(accuracy_score(y_test, rf_pred))
```

### 보고서 과제
1. voting, bagging, random forest, boosting, adaboost, gradient boost 의 특징 및 장단점을 스스로 정리해보기
2. Boosting의 advanced model 인 XGBoost, LightGBM, CatBoost에 대해 찾아보고 정리해보기

1-1 voting  
전체 dataset을 다양한 model로 학습시켜 최종 투표로 결과를 내는 방식  
장점: 다양한 model의 결과를 종합하여 안전성이 높다  
단점: 다양한 model을 전부 학습시켜야하여 시간이 오래걸린다  
  
1-2 bagging  
동일한 model로 dataset을 분류하여 학습시킨다. 각각의 학습된 model로 voting하여 결과를 내는 방식    
장점: 다양한 dataset에 대한 model을 만들기 때문에 과적합이 방지된다.  
단점: dataset을 분류할때 복원추출 방식으로 하기 때문에 학습되는 data가 독립이 아닐수 있어 비슷한 tree가 생길 수 있다.(이렇게 되면 전체모델의 분산이 증가한다고 한다.) 이러한 문제를 해결하기 위해 feature를 임의로 추출하여 분류하는 RandomForest 방식을 활용하기도 한다.

1-3 boosting  
dataset을 복원추출하여 나눈다. 나누어진 dataset을 순차적으로 학습하되 학습을 할때 이전학습의 오차를 보완하여 학습시키는 방식  
장점: 오차를 수정해나가는 방식으로 정확하다  
단점: 이상치에 취약하다
  
1-4 adaboosting  
잘못 분류된 데이터에 가중치를 높게 준어 다음 학습 때 데이터의 중요도가 올라간다.  
장점: 약분류기를 활용한 boosting으로 과적합을 피하기 쉽다  
단점: 역시 이상치에 취약하다  
  
1-5. gradient boosting  
학습 후의 잔차를 data로 다시 학습시키는 것을 반복하여 오차를 줄여나간다  
장점: 예측 성능이 우수하다  
단점: 순차적으로 학습하는과저이기 때문에 학습시간이 오래걸린다.  
  
2-1 XGBoost  
Extreme Gradient Boost의 약자로 기존의 순차적이었던 gradient boosting 과정을 병렬학습이 지원되도록 구현한 model이다.  
역시 gradient boosting이기 때문에 조정해야할 hyperparameter가 많지만 학습속도와 성능 모두 우수하다. 반대로 hyperparameter가 매우 중요하다. 
  
2-2 LightGBM  
light gradient boosting machine의 약어로 학습하는데 시간이 적고 메모리 사용량이 적다. 그렇기에 데이터사이언스 어플리케이션에도 적용된다고 한다. XGBoost를 비롯한 기존tree방식(level wise)과 다르게 깊이가 깊지않고 균형잡힌(leaf wise) 트리를 활용한다.  
적은 data로 학습하면 과적합 가능성이 높다고 한다.  
  
2-3 CatBoost  
residual boost를 활용한 boosting 으로 level wise tree 방식이다. 하지만 gradient boosting과는 다르게 잔차를 모아서 다음 dataset을 학습하고 그결과로 나온 잔차를 다시 학습하는 것을 반복한다. 이로서 과적합을 줄인다고는 하지만 왜 줄어드는지는 아직 정확하게는 모르겠다.  
수치형보다는 범주형 데이터에 좀더 적합하며, 기존의 XGBoost와 lightGBM이 hyperparameter에 민감하게 변한다는 단점을 보완했다고는 한다.
