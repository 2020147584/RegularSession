#### 2. 활성화 함수 (Activation function) 
- 정의: 입력을 받아서 활성화 또는 비활성화를 결정하는 데에 사용되는 함수. 
- 딥러닝 네트워크에서 노드에 입력된 값들을 비선형 함수에 통과시킨 후 다음 레이어로 전달하는데, 이 때 사용하는 함수를 활성화 함수(Activation Function)라고 한다.
- 선형 함수가 아니라 비선형 함수를 사용하는 이유는 딥러닝 모델의 레이어 층을 깊게 가져갈 수 있기 때문이다.
- 인공신경망에서 활성화 함수는 입력 데이터를 다음 레이어로 어떻게 출력하느냐를 결정하는 역할.
- 즉, 활성화 함수는 입력을 받아서 활성화 또는 비활성화를 결정하는 데에 사용되는 함수이다.

- 종류: 

1) 시그모이드 함수 (Sigmoid Function): 
- Logistic함수라고도 불리며 x값에 따라서 0 ~ 1의 값을 출력하는 S자형 함수로 범주형 분석 시에 사용되며 분류 값과 함께 해당 분류 값으로 분류될 확률을 함께 출력해준다는 특징이 있다. 

2) Tanh 함수 (Hyperbolic Tangent Function): 
- 쌍곡선 함수 중 하나로, sigmoid함수의 변형이다.
-  Tanh 함수는 함수의 중심점을 0으로 옮겨 sigmoid함수가 갖고 있던 최적화 과정에서 느려지는 문제를 해결했다. 
- But, 미분 함수에 대해 일정 값 이상에서 미분값이 소실되는 Vanishing Gradient 문제는 남아있다. 

3) ReLU 함수 (Rectified Linear Unit, 경사함수): 
- 가장 많이 사용되는 활성화 함수이다. 
- Sigmoid와 Tanh이 갖는 Gradient Vanishing 문제를 해결하기 위한 함수이다. 
- x가 x<0 범위에서는 y=0, x>0 범위에서는 y=x이다. 
- 때문에 x<0 범위에서 뉴런이 죽을 수 있다는 단점이 있다. 
- But still, Sigmoid, Tanh함수보다 학습이 빠르고, 연산비용이 적으며, 구현이 매우 간단하다는 특징이 있다. 

4) Leaky ReLU 함수: 
- ReLU 함수에서 Dying Relu(뉴런이 죽는 현상)가 발생하는 문제를 해결하기 위해 나온 함수이다. 
- x가 음수의 영역의 값에 대해 미분값이 되지 않는다는 점을 제외하면 ReLU의 특성을 동일하게 갖는다. 

5) PReLU 함수: 
- Leaky ReLU와 유사하나, 파라미터 a를 추가하여 x가 음수인 영역에서도 기울기를 학습한다. (= 음수에서 0의 값이 아니다.) 


6) ELU 함수(Exponential Linear Unit): 
- ReLU의 모든 장점을 포함하며 Dying ReLU문제를 해결했다. 
- 출력값이 거의 Zero centered와 가까우며, 일반적인 ReLU와 다르게 exp함수를 계산하는 비용이 발생한다. 

7) Maxout 함수:
- dying ReLU문제를 해결하지만, 계산 양이 많고 복잡하다.

#### 3. 손실 함수 (Loss function)
- 손실 함수 (Loss function)의 다른 명칭은 비용 함수(Cost Function)이다.
- 구축한 모델(model)에 입력값(input x)을 넣었을 때 산출되는 예측값(predicted y)이 실제 값(True y)과 얼마나 유사한지(얼마나 다른지)를 판단하는 함수이다.
- 예측값과 실제값의 차이를 loss라고 하며, 이 loss를 줄이는 방식으로 학습이 진행된다.  


##### [1] 회귀함수 손실함수 종류
- 종류: 

1) MAE (Mean Absolute Error, 평균 절대 오차):
- 예측값과 실제값의 차에 절댓값을 취하여 그 값들을 전부 더하고, 개수로 나누어 평균을 낸 값.
- 쉽게 예측값과 실제값의 차이를 파악할 수 있다.
- But, 절댓값을 취한 값이기 때문에 어떤 식으로 오차가 발생했는지 파악하기 어렵다. (양수 오차인지 음수 오차인지 등..)
- 또한 선형 + 절댓값 함수이기 때문에 (v형 그래프) 최적값에 가까워지더라도 이동거리가 일정하여, 최적 값에 수렴하기 어렵다. 

2) MSE (Mean Squared Error, 평균 제곱 오차):
- 가장 많이 사용되는 손실함수의 형태로 예측값과 실제값 사이의 평균을 제곱하여 평균을 낸 값이다. 
- 제곱 연산을 하기 때문에 차이가 커질수록 값이 뚜렷해진다. 
- 실제 정답에 대한 오차뿐만 아니라, 다른 오답들에 대한 정답률 오차 또한 포함하여 계산한다는 점이 특징적이다. 
- MSE는 MAE와 다르게 최적값에 가까워질 수록 이동거리가 변화하기 때문에 최적값 수렴에 용이하다. 
- But, 제곱을 하기 때문에 1미만의 값은 더 작아지고, 그 이상의 값은 더 커진다는 문제가 있다. 

3) RMSE (Root Mean Square Error, 평균 제곱근 오차) 
- MSE에 루트를 씌운 형태이다. 
- 제곱된 값(MSE)에 루트를 씌우기 때문에 값을 제곱해서 생기는 왜곡을 줄일 수 있다. 
- 또한, 제곱을 하지 않기 때문에 오차를 보다 직관적으로 보여준다. 

##### [2] 분류함수 손실함수 종류

- 종류: 
0) cross-entrophy방식 
- 실제 분포 q를 알지 못하는 상태에서, 모델링을 통해 구한 분포 p를 통해 실제분포(q)를 예측하는 방식이다. 
- 이 과정에서 실제 분포 q와 모델링된 분포 p가 모두 식에 포함되기 때문에 cross-entrophy라는 이름이 붙었다. 
- 실제값과 예측값이 맞는 경우 0으로 수렴하고, 값이 틀린 경우 발산하기 때문에 실제값과 예측값의 차이를 줄이기 위해 사용된다. 
- label값이 one-hot encoding인 경우에만 사용 가능하다. 

- *P.S.* one-hot encoding이란:
- Multinomial Classification에서 사용되는 인코딩 방법으로 출력 값의 형태가 정답 값 label인 경우 1 나머지는 0으로 표현되는 방식이다. ex) [1,0,0], [0, 1, 0] , [0, 0, 1] 
- 더미 변수 (Dummy Variables)를 만들어 분류하는 방식으로 코딩할 수 있을 것으로 보인다.

#### 4. 경사 하강법 (Gradient descent)
- 정의: 함수 값이 낮아지는 방향으로 독립 변수 값을 변형시켜가면서 최종적으로는 최소 함수 값을 갖도록 하는 독립 변수 값을 찾는 방법 
- 비유: 앞이 보이지 않는 안개가 낀 산을 내려올 때는 모든 방향으로 산을 더듬어가며 산의 높이가 가장 낮아지는 방향으로 한 발씩 내딛어갈 수 있다.

목적과 사용 이유:
- 우리가 주로 실제 분석에서 맞딱드리게 되는 함수들은 닫힌 형태(closed form)가 아니거나 함수의 형태가 복잡해 (가령, 비선형함수) 미분계수와 그 근을 계산하기 어려운 경우가 많다. 
- 실제 미분계수를 계산하는 과정을 컴퓨터로 구현하는 것에 비해 gradient descent는 컴퓨터로 비교적 쉽게 구현할 수 있기 때문이다.
- 데이터 양이 매우 큰 경우 gradient descent와 같은 iterative한 방법을 통해 해를 구하면 계산량 측면에서 더 효율적으로 해를 구할 수 있다.

- 학습률이 너무 작은 경우 학습 속도가 굉장히 작다. 
- 학습률이 너무 큰 경우에는 최저점을 찾지 못하고 이탈할 수 있다.

### 신경망 모델 구축

신경망 클래스 이름은 Net으로 하여 3개의 레이어를 가진 모델을 만드세요.  
노드의 개수와 활성화 함수는 자유입니다.  
출력층 노드의 개수는 10개입니다.  
또한, 출력층에 Softmax 함수를 적용할 필요가 없습니다.

(hint. forward시 첫 번째 레이어를 통과하기 전에 784차원의 텐서로 차원을 변환 해야합니다.)

```python
############### Net 신경망 생성 ##############

class Net(nn.Module):
    def __init__(self):
       super(Net, self).__init__()
       self.layer1 = nn.Linear(784, 526)
       self.layer2 = nn.Linear(526, 268)
       self.layer3 = nn.Linear(268, 10)
       self.relu = nn.ReLU()

    def forward(self, x):
      x = x.view([-1, 784])
      out = self.layer1(x)
      out = self.relu(out)
      out = self.layer2(out)
      out = self.relu(out)
      out = self.layer3(out)
      
      return out
```

```python
def train(model, train_loader, optimizer):
    model.train()

    ######## 배치 당 loss 값을 담을 리스트 생성 #########
    batch_losses = []

    for data, target in train_loader:
        ######### 옵티마이저의 기울기 초기화 ############
        optimizer.zero_grad()

        ######### 모델에서 예측값 산출 ############
        output = model.forward(data)

        ######### loss 계산 ############
        # 정답 데이터와의 Cross Entropy Loss 계산
        # 그리고 이 loss를 배치 당 loss로 보관합니다.
        loss = criterion(output, target)
        batch_losses.append(loss)
        


        ######### 기울기 계산 ########
        loss.backward()

        ######## 가중치 갱신 ########
        optimizer.step()
    
    ######## 배치 당 평균 loss 계산 ############
    avg_loss = sum(batch_losses) / len(batch_losses)
    
    return avg_loss
```

모델을 테스트하는 과정의 함수도 생성합니다.  
주석으로 지시하는 부분의 코드를 작성해주세요.  

```python
def evaluate(model, test_loader):
    # 모델을 평가 모드로 전환
    model.eval()

    ######## 배치 당 loss 값을 담을 리스트 생성 #########
    batch_losses = []

    # 실제 모델의 예측이 정답과 맞은 횟수 = correct
    correct = 0

    ####### 평가 과정에서는 기울기를 계산하지 않습니다. #########
    ####### 해당하는 코드를 작성해주세요. ############
    with torch.no_grad(): 
        for data, target in test_loader:
            # 예측값 생성
            output = model(data)

            ######### loss 계산 ############
            # 정답 데이터와의 Cross Entropy Loss 계산
            # 그리고 이 loss를 배치 당 loss로 보관합니다.
            loss = criterion(output, target)
            batch_losses.append(loss)
            
            
            # Accuracy 계산
            # 가장 큰 값을 가진 클래스가 모델의 예측입니다.
            # 예측 클래스(pred)과 정답 클래스를 비교하여 일치할 경우 correct에 1을 더합니다.
            pred = output.max(1, keepdim=True)[1]

            # eq() 함수는 값이 일치하면 1을, 아니면 0을 출력.
            correct += pred.eq(target.view_as(pred)).sum().item()

    ######## 배치 당 평균 loss 계산 ############
    avg_loss = sum(batch_losses) / len(batch_losses)

    #정확도 계산
    accuracy = 100. * correct / len(test_loader.dataset)

    return avg_loss, accuracy
```

코드를 실행시켜 학습을 진행합니다.  
다소 시간이 걸릴 수 있습니다.
